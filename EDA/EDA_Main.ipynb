# Overview

### This file shows the EDA. The annotated images are first cropped and stored in folders according to their class.  The pixel values for the red, blue and green channels, and greyscale images was normalised for each image and aggregated for each class (Day 0 -7). This data was then plotted using a box pot to examine if there were any patterns in pixel values as the fish deteriorates. The observations from the cropped images and colour channels are recorded below.

```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image  # You may need to install the Pillow library using `pip install Pillow`
from matplotlib import gridspec

# Set the path to the directory containing your image folders
base_path = "./cropped_folder/"

# Get the list of folders in the base path
folders = os.listdir(base_path)

# Set the number of images to randomly pick from each folder
num_images_per_folder = 10

# Set the common size for resizing
common_width = 100
common_height = 100

# Set up the plot with gridspec
fig = plt.figure(figsize=(12, 12))
gs = gridspec.GridSpec(len(folders), num_images_per_folder, width_ratios=[1] * num_images_per_folder)

# Loop through each folder
for i, folder in enumerate(folders):
    folder_path = os.path.join(base_path, folder)
    
    # Get a list of image files in the folder
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
    
    # Randomly pick 10 images
    selected_images = random.sample(image_files, num_images_per_folder)
    
    # Display the images in the plot using gridspec
    for j, image_file in enumerate(selected_images):
        image_path = os.path.join(folder_path, image_file)
        
        # Open the image and resize it
        img = Image.open(image_path)
        img = img.resize((common_width, common_height))
        
        # Display the resized image using gridspec
        ax = plt.subplot(gs[i, j])
        ax.imshow(img)
        ax.axis('off')

# Label the rows with the folder names on the left axis
for i, folder in enumerate(folders):
    ax = plt.subplot(gs[i, 0])
    ax.text(0, 0.5, folder+'  ', rotation=0, ha='right', va='center', transform=ax.transAxes)

# Adjust layout for better visualization
plt.tight_layout(w_pad=0.1, h_pad=0.05)
plt.show()
```

## Observation from Fish Eyes

### Generally, on day 0 the fish eyes are mostly back with a golden perimeter. From days 1-2 the fish eyes become, foggy and white but still translucent. In day 4 the fish eyes turn dry and less translucent.  Usually after day 4, the fish eyes will then tend to become pinkish and red. 

### However, during the experiment, the fishes were observed to deteriorate at different rates. One of the fish eyes became foggy on day 0 itself, while the other 2 remained bright and black, whereas another fish was observed to deteriorate very slowly. On day 2 its eyes were still mostly black and a little foggy while the eyes of the other 2 were mostly white. 


```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def extract_red_channel(image_path):
    # Read the image using OpenCV
    image = cv2.imread(image_path)

    # Extract the red channel
    red_channel = image[:, :, 2]

    return red_channel

def plot_boxplot(ax, data, channel_names, title, position):
    # Normalize the pixel intensities for each image
    normalized_data = [[pixel / 255.0 for pixel in image] for image in data]

    # Concatenate flattened arrays for each channel
    concatenated_data = [np.concatenate(image) for image in normalized_data]

    # Set box properties for red color
    box_props = dict(facecolor='red', edgecolor='red', linewidth=2)

    # Set whisker properties for black color
    whisker_props = dict(color='black', linewidth=2)

    # Set median properties for black color
    median_props = dict(color='black', linewidth=2)

    # Plot box plots for each color channel
    bplot = ax.boxplot(concatenated_data, labels=channel_names, whis=100, positions=[position],
                       boxprops=dict(facecolor='red', edgecolor='red', linewidth=2),
                       whiskerprops=whisker_props, medianprops=median_props, patch_artist=True)
    
    # Set the face color for the boxes
    for patch in bplot['boxes']:
        patch.set_facecolor('red')
    
    ax.set_title(title)
    ax.set_xlabel('Red Channels')
    ax.set_ylabel('Normalized Pixel Intensity')

def main(folders):
    # Create a single subplot for all data
    fig, ax = plt.subplots(figsize=(15, 10))

    for i, folder_path in enumerate(folders):
        # Get a list of all image files in the folder
        image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        # Initialize lists to store red channel data
        red_channel_data = []

        # Iterate over each image and extract red channel data
        for image_file in image_files:
            image_path = os.path.join(folder_path, image_file)
            red_channel = extract_red_channel(image_path)

            # Normalize the red channel and flatten
            flattened_data = (red_channel / 255.0).flatten()

            # Append normalized red channel data to the list
            red_channel_data.append(flattened_data)

        # Plot box plots for each folder
        plot_boxplot(ax, [red_channel_data], ['Red Channel'], f'Combined Plot', position=i)

    # Label x-axis with folder names
    ax.set_xticks(range(len(folders)))
    ax.set_xticklabels([f"Day {i}" for i in range(len(folders))])

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    folders = ["./cropped_folder/Day_0/", "./cropped_folder/Day_1/", "./cropped_folder/Day_2/",
               "./cropped_folder/Day_3/", "./cropped_folder/Day_4/", "./cropped_folder/Day_5/",
               "./cropped_folder/Day_6/", "./cropped_folder/Day_7/"]
    main(folders)
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def extract_grayscale(image_path):
    # Read the image using OpenCV
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    return grayscale_image

def plot_boxplot(ax, data, channel_names, title, position, box_color):
    # Normalize the pixel intensities for each image
    normalized_data = [[pixel / 255.0 for pixel in image] for image in data]

    # Concatenate flattened arrays for each channel
    concatenated_data = [np.concatenate(image) for image in normalized_data]

    # Plot box plots for the grayscale channel
    bplot = ax.boxplot(concatenated_data, labels=channel_names, whis=100, positions=[position], patch_artist=True)
    
    # Set the box color for the boxes
    for patch in bplot['boxes']:
        patch.set_facecolor(box_color)
        patch.set_edgecolor('black')  # Set the edge color to black
        patch.set_linewidth(2)  # Set the line width to 2

    # Set whisker properties for black color
    whisker_props = dict(color='black', linewidth=2)

    # Set median properties for black color
    median_props = dict(color='black', linewidth=2)

    for key in ('whiskers', 'medians', 'caps'):
        for element in bplot[key]:
            plt.setp(element, color='black', linewidth=2)  # Set the line width to 2

    ax.set_title(title)
    ax.set_xlabel('Grayscale Channel')
    ax.set_ylabel('Normalized Pixel Intensity')

def main(folders):
    # Create a single subplot for all data
    fig, ax = plt.subplots(figsize=(15, 10))

    for i, folder_path in enumerate(folders):
        # Get a list of all image files in the folder
        image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        # Initialize lists to store grayscale data
        grayscale_data = []

        # Iterate over each image and extract grayscale data
        for image_file in image_files:
            image_path = os.path.join(folder_path, image_file)
            grayscale_image = extract_grayscale(image_path)

            # Normalize the grayscale image and flatten
            flattened_data = (grayscale_image / 255.0).flatten()

            # Append normalized grayscale data to the list
            grayscale_data.append(flattened_data)

        # Plot box plots for each folder
        plot_boxplot(ax, [grayscale_data], ['Grayscale'], f'Combined Plot', position=i, box_color='gray')

    # Label x-axis with folder names
    ax.set_xticks(range(len(folders)))
    ax.set_xticklabels([f"Day {i}" for i in range(len(folders))])

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    folders = ["./cropped_folder/Day_0/", "./cropped_folder/Day_1/", "./cropped_folder/Day_2/",
               "./cropped_folder/Day_3/", "./cropped_folder/Day_4/", "./cropped_folder/Day_5/",
               "./cropped_folder/Day_6/", "./cropped_folder/Day_7/"]
    main(folders)
```

## Observations from Colour Channels

### Day 0 was observed to usually have low magnitude across all colour channels, and day 4 had the highest magnitude. Based on the grey scale box plot, and cropped images, the fish eyes were mostly black on day 0 and white on day 4. Since white is a combination of red, green and blue, at maximum pixel values, and black is represented by near zero pixel values, this could be the reason why day 4 had the highest magnitude across all three colour channels, and day 0 had the lowest.
